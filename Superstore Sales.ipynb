```python
import pandas as pd

df = pd.read_csv('SampleSuperstore.csv')
print('Data loaded successfully.')
print('First 5 rows of the dataset:')
print(df.head())
print('\
Data Types and Missing Values:')
print(df.info())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values)

# Check for duplicates
print('Number of duplicate rows:', df.duplicated().sum())

# Check for inconsistencies in categorical data
for column in df.select_dtypes(include=['object']).columns:
    print(f'Inconsistencies in {column}:', df[column].apply(lambda x: x.strip()).nunique() - df[column].nunique())

# Convert data types if necessary
# Checking data types that might need conversion
print('Data types before conversion:')
print(df.dtypes)

# Converting 'Postal Code' from float to int
if df['Postal Code'].dtype != 'int':
    df['Postal Code'] = df['Postal Code'].astype(int)
    print('Postal Code converted to int.')

print('Data types after conversion:')
print(df.dtypes)
```

```python
# Remove duplicate rows
initial_row_count = len(df)
df = df.drop_duplicates()
final_row_count = len(df)

# Display the result
print('Initial number of rows:', initial_row_count)
print('Number of rows after removing duplicates:', final_row_count)
```

```python
print('Summary Statistics for Numerical Columns:')
print(df.describe())

print('Unique Values in Categorical Columns:')
for column in df.select_dtypes(include=['object']).columns:
    print(f'Unique values in {column}:', df[column].nunique())
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Remove duplicates
initial_row_count = len(df)
df = df.drop_duplicates()
final_row_count = len(df)

# Sales by State
plt.figure(figsize=(15, 8), facecolor='white')
sns.barplot(x='State', y='Sales', data=df, estimator=sum, ci=None)
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Profit by State
plt.figure(figsize=(15, 8), facecolor='white')
sns.barplot(x='State', y='Profit', data=df, estimator=sum, ci=None)
plt.title('Total Profit by State')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Correlation matrix
plt.figure(figsize=(10, 8), facecolor='white')
corr = df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Additional Descriptive Statistics
skewness = df[['Sales', 'Profit', 'Quantity', 'Discount']].skew()
kurtosis = df[['Sales', 'Profit', 'Quantity', 'Discount']].kurtosis()
range_sales = df['Sales'].max() - df['Sales'].min()
iqr_sales = df['Sales'].quantile(0.75) - df['Sales'].quantile(0.25)
sum_sales = df['Sales'].sum()
unique_states = df['State'].nunique()

(skewness, kurtosis, range_sales, iqr_sales, sum_sales, unique_states)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'SampleSuperstore.csv'
df = pd.read_csv(file_path, encoding='ascii')

# Calculate Profit Margin as Profit divided by Sales
# Adding a small constant to avoid division by zero
df['Profit Margin'] = df['Profit'] / (df['Sales'] + 1e-6)

# Plotting the distribution of Profit Margin
plt.figure(figsize=(10, 6), facecolor='white')
sns.histplot(df['Profit Margin'], kde=True, color='purple')
plt.title('Distribution of Profit Margin')
plt.xlabel('Profit Margin')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Displaying the head of the dataframe with the new column
print(df[['Profit Margin']].head())
```

```python
# Calculating the average profit margin for each sub-category
subcat_profit_margin = df.groupby('Sub-Category')['Profit Margin'].mean().sort_values(ascending=False)

# Displaying the top 5 sub-categories with the highest profit margin
top_5_subcat_profit_margin = subcat_profit_margin.head(5)
print(top_5_subcat_profit_margin)
```

```python
# Calculating the average profit margin for each sub-category
subcat_profit_margin = df.groupby('Sub-Category')['Profit Margin'].mean().sort_values(ascending=True)

# Displaying the top 5 sub-categories with the lowest profit margin
top_5_subcat_lowest_profit_margin = subcat_profit_margin.head(5)
print(top_5_subcat_lowest_profit_margin)
```

```python
# Grouping the data by Category to sum the sales
category_sales = df.groupby('Category').agg({'Sales':'sum'}).reset_index()

# Displaying the sales breakdown by category
category_sales
```

```python
# Finding the top-selling products in each category
top_products_by_category = df.groupby(['Category', 'Sub-Category']).agg({'Sales':'sum'}).sort_values(by=['Category', 'Sales'], ascending=[True, False]).groupby('Category').head(1).reset_index()

# Displaying the top-selling products by category
top_products_by_category
```

```python
# Grouping the data by Region to sum the sales
region_sales = df.groupby('Region').agg({'Sales':'sum'}).reset_index()

# Displaying the sales breakdown by region
region_sales
```

```python
# Finding the top-selling products in each region
top_products_by_region = df.groupby(['Region', 'Sub-Category']).agg({'Sales':'sum'}).sort_values(by=['Region', 'Sales'], ascending=[True, False]).groupby('Region').head(1).reset_index()

# Displaying the top-selling products by region
top_products_by_region
```

```python
import pandas as pd

df = pd.read_csv('SampleSuperstore.csv', encoding='ascii')

# Grouping data by 'State' and 'City' to analyze geographic purchasing behavior
df_geo = df.groupby(['State', 'City']).agg({'Sales': 'sum', 'Profit': 'sum', 'Quantity': 'sum'}).reset_index()

# Sorting the data by Sales to see top purchasing cities
df_geo_sorted = df_geo.sort_values(by='Sales', ascending=False).head(10)

print(df_geo_sorted)
```
